```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('pk_candidate_data.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df['election_date'] = pd.to_datetime(df['election_date'])
df['election_date'] = df['election_date'].dt.year
```

```python
ConstituencyDf = df[['election_date', 'constituency_number', 'constituency_name', 'province', 'assembly', 'voter_reg', 'validated_votes', 'votes_disq']]
CandidateDf = df[['candidateID', 'candidate_name', 'candidate_party']]
edgeDf = df[['election_date', 'constituency_number', 'candidateID', 'candidate_votes', 'candidate_share', 'candidate_rank', 'outcome']]
```

```python
ConstituencyDf = ConstituencyDf.dropna()
CandidateDf = CandidateDf.dropna()
edgeDf = edgeDf.dropna()
```

```python
ConstituencyDf.to_csv('Constituency.csv', index=False)
CandidateDf.to_csv('Candidate.csv', index=False)
edgeDf.to_csv('edge.csv', index=False)
```
